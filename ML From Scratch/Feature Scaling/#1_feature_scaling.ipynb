```python
import numpy as np 
```

```python
X = np.random.randint(10, 3000, size=(100,8)).astype('float')
X
```

### First Strategy: Divide each value by its maximum value. This will result in a range 0 <= x_i => 1

```python
n = X.shape[1]
# X_max = np.zeros((n,))
X_copy = np.copy(X)

for i in range(n):
    max_val = np.max(X_copy[:, i])
    X_copy[:, i] = X_copy[:, i] / max_val

X_copy
```

### Second Strategy is Min-Max Normalization (0 - 1 common range)

```python
X_copy = np.copy(X)

for i in range(X_copy.shape[1]):
    X_max = np.max(X_copy[:, i])
    X_min = np.min(X_copy[:, i])

    X_copy[:, i] = (X_copy[:, i] - X_min) / (X_max - X_min)

X_copy
```

### Third Strategy is Standard Scaling (Z-score normalization) (mean = 0, std = 1)

```python
X_copy = np.copy(X)

for i in range(X_copy.shape[1]):
    X_copy[:, i] = (X_copy[:, i] - np.mean(X_copy[:, i])) / np.std(X_copy[:, i])

X_copy
```

```python
print(f"Mean: {np.mean(X_copy[:, 0])}, Std: {X_copy[:, 0]}")
```
