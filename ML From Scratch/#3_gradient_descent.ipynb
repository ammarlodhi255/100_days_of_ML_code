```python
import numpy as np 
```

```python
def f_wb(X, w, b):
    return X * w + b 
```

```python
def compute_cost(Y_hat, Y, w, b):
    m = Y.shape[0]
    cost = (1/(2*m)) * np.sum((Y_hat - Y) ** 2)
    return cost
```

```python
def derivative_w(Y_hat, X, Y):
    m = Y.shape[0]
    return (1/m) * np.sum((Y_hat - Y) * X)

def derivative_b(Y_hat, Y):
    m = Y.shape[0]
    return (1/m) * np.sum(Y_hat - Y)
```

```python
def gradient_descent(X, Y, w, b, lr=0.1, iterations=100):
    min_cost = float('inf')
    min_cost_params = {'w': 0, 'b': 0}

    for i in range(iterations):
        Y_hat = f_wb(X, w, b)

        temp_w = w - lr * derivative_w(Y_hat, X, Y)
        temp_b = b - lr * derivative_b(Y_hat, Y)
        w = temp_w 
        b = temp_b 

        cost = compute_cost(Y_hat, Y, w, b)
        print(f"w={w}, b={b}")
        print(f"Current Cost: {cost}")

        if cost < min_cost:
            min_cost = cost 
            min_cost_params['w'] = w
            min_cost_params['b'] = b


    print(f'Best Min Cost: {min_cost}, {min_cost_params}')
```

```python
w = 100
b = 100

X = np.array([1.0, 2.0])
Y = np.array([300, 500])

gradient_descent(X,Y, w, b, iterations=1000)
```
