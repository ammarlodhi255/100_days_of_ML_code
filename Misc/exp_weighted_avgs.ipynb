```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
data = np.random.random_sample((1000,))
data
```

```python
def polynomial_function(x):
    return 0.5 * x**3 - 2 * x**2 + 3 * x + 5

# Generate random x values
x_values = np.random.uniform(-5, 5, 100)

# Generate corresponding y values with some random noise
y_values = polynomial_function(x_values) + np.random.normal(0, 5, 100)

# Plot the generated data points
plt.scatter(x_values, y_values, color='blue', label='Random Data Points')
```

```python
def compute_exp_weighted_avg(beta, data):
    weighted_avgs = np.zeros((data.shape[0],))

    for i in range(len(data)):
        if i == 0:
            v = (1-beta)*data[i]
        else: v = beta*weighted_avgs[i-1] + (1-beta)*data[i]

        weighted_avgs[i] = v

    return weighted_avgs
```

```python
new_values = compute_exp_weighted_avg(0.2, y_values)
```

```python
plt.scatter(x_values, new_values, color='blue', label='Random Data Points')
```
